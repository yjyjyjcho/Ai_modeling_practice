# **케라스 순차형 모델 구조**

```


from keras.models import Sequential
from keras.layers import Dense, Activation

#객체생성
model = Sequential()

#레이어 쌓기
model.add(Dense(32, activation = 'relu', input_dim=100))
model.add(Dense(1, activation='sigmoid'))

#model.add(layer_name)

#학습과정 구성
model.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['accuracy'])

#데이터 생성
import numpy as np
data = np.random.random((1000,100))
labels = np.random.randint(2, size=(1000,1))

#모델 훈련
model.fit(data, labels, epochs=10, batch_size=32)

#모델 정보
model.summary()
```

# **케라스 함수형 API 모델 구조**

```
from keras.layers import Input, Dense
from keras.models import Model

#tensor(입력 데이터 모양) 객체 생성
#입력 레이어는 입력 데이터의 차원을 나타내는 튜플인 shape 인수를 받는다.

InputTensor = Input(shape=(100,))

#레이어 정의
H1 = Dense(10, activation='relu')(InputTensor) #입력 레이어의 출력을 dense레이어 입력을 연결
H2 = Dense(20, activation='relu')(H1) #dense레이어 출력(H1)을 다른 dense레이어(H2)의 입력으로 연결
Output = Dense(1, activation='softmax')(H2)

model = Model(inputs=InputTensor, outputs = Output)

model.summary()
```
