## **perceptron**

```
# perceptron.py
# Perceptron 알고리즘 구현 파일
import numpy as np

class perceptron:
  #생성자
  #thresholds : 임계값, 계산된 예측 값을 비교하는 값
  # eta : 학습률
  # n_iter : 학습 횟수
    def __init__(self, thresholds=0.0, eta=0.01, n_iter=10):
      self.thresholds = thresholds
      self.eta = eta
      self.n_iter = n_iter #몇번 계산할지(학습횟수)

    #학습 함수
    #X : 입력데이터 y : 결과데이터
    def fit(self, X, y):
    #가중치를 담을 행렬 생성
        self.w_ = np.zeros(1 + X.shape[1])
        #예측값과 실제값을 비교
        #다른 예측값의 갯수를 담는 리스트
        self.errors_ = []

        #지정된 학습 횟수 만큼 반복
        for _ in range(self.n_iter):  #_ : 아무것도 안담는다는 뜻, 값을 받는데 따로 저장을 안하려고 할 때
            #예측값과 실제값이 다른 개수를 담는 변수
            errors = 0
            temp1 = zip(X, y) # zip : X, y 를 튜플로 하나로 묶어줌
            for xi, target in temp1:
                a1 = self.predict(xi)
                if target != a1: 
                    update = self.eta * (target - a1) #에러펑션은 간단히 구현했음
                    self.w_[1:] += update * xi #가중치들
                    self.w_[0] += update #바이어스
                    errors += int(update != 0.0)
            self.errors_.append(errors)
            print(self.w_)
            
    #가중치 * 입력값의 총합을 계산
    # X : 입력값
    def net_input(self, X):
        a1 = np.dot(X, self.w_[1:] + self.w_[0])
        return a1

    #예측된 결과를 가지고 판단
    #X : 입력값 배열
    def predict(self, X):
        a2 = np.where(self.net_input(X) > self.thresholds, 1, -1)
        return a2
```

```
import numpy as np
from perceptron import perceptron
from time import time
import pickle

def step1_learning():
    #학습과 테스트를 위해 사용 할 데이터
    X = np.array([[0,0], [0,1], [1,0], [1,1]])
    y = np.array([-1, -1, -1, 1])

    #퍼셉트론 객체 인스턴스 생성
    ppn = perceptron(eta=0.1)

    #학습
    stime = time()
    ppn.fit(X, y)
    etime = time()
    print('학습에 걸린 시간 : ', (etime - stime)) 
    print('학습 중 오차가 난 갯수 : ', ppn.errors_)

    #학습이 완료된 객체를 파일로 저장
    with open('./perceptron.dat', 'wb') as fp:
        pickle.dump(ppn, fp)
    print('학습 완료')


def step2_using():
    #파일로부터 객체를 복원
    with open('./perceptron.dat', 'rb') as fp:
        ppn = pickle.load(fp)
    while True:
        a1 = input('첫번째 2진수를 입력하세요 : ')
        a2 = input('두번째 2진수를 입력하세요 : ')
        X = np.array([int(a1), int(a2)])
        result = ppn.predict(X)
        if result == 1:
            print('결과 : 1')
        else :
            print('결과 : 0')


if __name__ == "__main__":
    step1_learning()
    #step2_using()
```
